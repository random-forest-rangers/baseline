```python
import pandas as pd
import glob

# Collect file paths
train_input_files = sorted(glob.glob("/kaggle/input/nfl-big-data-bowl-2026-prediction/train/input_*.csv"))
train_output_files = sorted(glob.glob("/kaggle/input/nfl-big-data-bowl-2026-prediction/train/output_*.csv"))

# Sanity check: the same number of input and output files?
print(f"Input files: {len(train_input_files)}")
print(f"Output files: {len(train_output_files)}")
print("First 3 input files:", train_input_files[:3])
print("First 3 output files:", train_output_files[:3])

# Load and concatenate all inputs/outputs
train_inputs = pd.concat([pd.read_csv(f) for f in train_input_files], ignore_index=True)
train_outputs = pd.concat([pd.read_csv(f) for f in train_output_files], ignore_index=True)

# Load test files
test_inputs = pd.read_csv("/kaggle/input/nfl-big-data-bowl-2026-prediction/test_input.csv")
test_meta   = pd.read_csv("/kaggle/input/nfl-big-data-bowl-2026-prediction/test.csv")

# Print shapes and peek
print("train_inputs:", train_inputs.shape)
print("train_outputs:", train_outputs.shape)
print("test_inputs:", test_inputs.shape)
print("test_meta:", test_meta.shape)

train_inputs.head()
```

```python
import pandas as pd

# Load just week 1
train_in = pd.read_csv("/kaggle/input/nfl-big-data-bowl-2026-prediction/train/input_2023_w01.csv")
train_out = pd.read_csv("/kaggle/input/nfl-big-data-bowl-2026-prediction/train/output_2023_w01.csv")

# Merge week 1 only
train_week1 = pd.merge(train_in, train_out, on=["game_id", "play_id"], how="inner")

print(train_week1.shape)
train_week1.head()
```

```python
def add_features(df):
    # Speed magnitude (using s = speed, a = acceleration if present)
    df["speed_mag"] = (df["s"]**2 + df["a"]**2)**0.5
    
    # Direction in radians (convert degrees to radians for modeling)
    df["dir_rad"] = np.deg2rad(df["dir"])
    
    return df

# Apply to training and test inputs
train_week1 = add_features(train_week1)
test_inputs = add_features(test_inputs)

train_week1.head()
```

```python
# Keep only numeric columns for modeling
X = train_week1.drop(columns=["ball_land_x", "ball_land_y"])
X = X.select_dtypes(include=["int64", "float64", "bool"])

y_x = train_week1["ball_land_x"]
y_y = train_week1["ball_land_y"]

print("Feature columns:", X.columns.tolist()[:10], "...")  # look at first few
```

```python
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Targets
y_x = train_week1["ball_land_x"]
y_y = train_week1["ball_land_y"]

# Train/validation split
X_train, X_val, yx_train, yx_val = train_test_split(X, y_x, test_size=0.1, random_state=42)
_, _, yy_train, yy_val = train_test_split(X, y_y, test_size=0.1, random_state=42)

# Train two models
model_x = LGBMRegressor(n_estimators=200, learning_rate=0.05, max_depth=7)
model_y = LGBMRegressor(n_estimators=200, learning_rate=0.05, max_depth=7)

model_x.fit(X_train, yx_train)
model_y.fit(X_train, yy_train)

# Validation scores
pred_x = model_x.predict(X_val)
pred_y = model_y.predict(X_val)

print("Validation MSE (x):", mean_squared_error(yx_val, pred_x))
print("Validation MSE (y):", mean_squared_error(yy_val, pred_y))
```

```python
train_features = X.columns
X_test = test_inputs.reindex(columns=train_features, fill_value=0)

print("Train features:", len(train_features))
print("Test features:", X_test.shape[1])
```

```python
# Predict frame-level
test_preds_x = model_x.predict(X_test)
test_preds_y = model_y.predict(X_test)

# Attach predictions to play_id
preds = pd.DataFrame({
    "play_id": test_inputs["play_id"],
    "ball_land_x": test_preds_x,
    "ball_land_y": test_preds_y
})

# Aggregate to play-level (mean prediction per play_id)
agg_preds = preds.groupby("play_id", as_index=False).mean()

# Merge with ALL play_ids in test_meta
submission = test_meta[["play_id"]].merge(agg_preds, on="play_id", how="left")

# Fill missing predictions with global mean (safe baseline strategy)
submission["ball_land_x"] = submission["ball_land_x"].fillna(agg_preds["ball_land_x"].mean())
submission["ball_land_y"] = submission["ball_land_y"].fillna(agg_preds["ball_land_y"].mean())

# Save submission
submission.to_csv("submission.csv", index=False)
print("Submission file created:", submission.shape)
submission.head()
```
